```python
import copy
def printroom(room):
    for i in range(len(room)):
        for j in range(len(room[0])):
            print(room[i][j],end=" ")
        print("")
    print("\n")

def st1(room):
    m,n = len(room),len(room[0])
    newroom = copy.deepcopy(room)
    if room[0][0] == 1:
        newroom[0][0]=0
    for i in range(m):
        for j in range(1,n):
            if room[i][j]==1 and room[i][j-1]==0:
                newroom[i][j-1],newroom[i][j] = newroom[i][j],newroom[i][j-1]
    return newroom

def st2(room):
    m,n = len(room),len(room[0])
    newroom = copy.deepcopy(room)
    if room[0][0] == 1:
        newroom[0][0]=0
    for i in range(1,m):
        for j in range(n):
            if room[i][j]==1 and room[i-1][j]==0:
                newroom[i-1][j],newroom[i][j] = newroom[i][j],newroom[i-1][j]
    return newroom

def aaa(room,s):
    for t in range(s):
        room = st1(room)
        room = st2(room)
    return room

room = [[0,1,0,1,0],[1,0,0,1,1],[1,1,0,1,0],[0,0,1,0,0]]
aaa(room,6)
```

```python
import copy
def step(room,tup):
    newroom = copy.deepcopy(room)
    if room[0][0]:
        newroom[0][0]=0
    for i in range(len(room))[tup[0]:]:
        for j in range(len(room[0]))[tup[1]:]:
            if room[i][j] and not room[i-tup[0]][j-tup[1]]:
                newroom[i-tup[0]][j-tup[1]],newroom[i][j] = newroom[i][j],newroom[i-tup[0]][j-tup[1]]
    return newroom
def ex5_7(room,s):
    for _ in range(s):
        room = step(step(room,(0,1)),(1,0))
    return room
room = [[0,1,0,1,0],[1,0,0,1,1],[1,1,0,1,0],[0,0,1,0,0]]
ex5_7(room,6)
```
